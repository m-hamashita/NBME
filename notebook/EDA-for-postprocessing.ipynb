## feature_text の uni-gram, bi-gram を取ってきて、 annotation rate(sum(annotation)/sum(pn_history)) が大きいやつに対して hard labeling する試み

```python
# feature_texts = train["feature_text"].unique()
feature_dic = dict(zip(features["feature_num"], features["feature_text"]))
```

```python
train["feature_text_list"] = train["feature_text"].apply(lambda x: x.split("-OR-")).apply(lambda x: [x1.split("-") for x1 in x])
```

```python
def ngram(feature_text_lists, n: int):
    return [" ".join(feature_text_list[k:k+n]) for feature_text_list in feature_text_lists for k in range(len(feature_text_list)-n+1)]
```

```python
train["feature_text_unigram"] = train["feature_text_list"].apply(ngram, n=1)
train["feature_text_bigram"] = train["feature_text_list"].apply(ngram, n=2)
```

```python
train["feature_text_bigram"]
```

```python
# heart_pounding["annotation"]
```

```python
feature_num_list_unigram = {feature_num: [ngram([feature.lower().split("-")], 1) for feature in feature_text.split("-OR-")] for feature_num, feature_text in feature_dic.items()}
feature_num_list_bigram = {feature_num: [ngram([feature.lower().split("-")], 2) for feature in feature_text.split("-OR-")] for feature_num, feature_text in feature_dic.items()}
```

```python
feature_texts = train["feature_text"].unique()
```

```python
feature_text_list_unigram = {feature_text: [ngram([feature.lower().split("-")], 1) for feature in feature_text.split("-OR-")] for feature_text in feature_texts}
feature_text_list_bigram = {feature_text: [ngram([feature.lower().split("-")], 2) for feature in feature_text.split("-OR-")] for feature_text in feature_texts}
```

```python
feature_text_list_unigram
```

```python
from tqdm import tqdm
annotation_rate_dic = {}
threshold = 0.8
for feature_num in feature_num_list_unigram.keys():
    annotation_rate_dic[feature_num] = []
for feature_num, word_lists in feature_num_list_unigram.items():
    for word_list in word_lists:
        for word in word_list:
            annotation_count = sum(
                train[train["feature_num"] == feature_num]["annotation"].apply(
                    lambda x: word in " ".join(x).lower()
                )
            )
            pn_history_count = sum(
                train[train["feature_num"] == feature_num]["pn_history"].apply(
                    lambda x: word in x.lower()
                )
            )
            if pn_history_count != 0:
                annotation_rate = annotation_count / pn_history_count
                if annotation_rate > 0.8:
                    annotation_rate_dic[feature_num].append(
                        {
                            "word": word,
                            "annotation_rate": annotation_rate,
                            "pn_history_count": pn_history_count,
                        }
                    )
            else:
                print(f"{feature_num}, {word}: pn_history_count=0")
```

```python
def match_annotation_rate(feature_num_list, threshold=0.8):
    annotation_rate_dic = {}
    for feature_num, word_lists in feature_num_list.items():
        annotation_rate_dic[feature_num] = []
        for word_list in word_lists:
            for word in word_list:
                annotation_count = sum(
                    train[train["feature_num"] == feature_num]["annotation"].apply(
                        lambda x: word in " ".join(x).lower()
                    )
                )
                pn_history_count = sum(
                    train[train["feature_num"] == feature_num]["pn_history"].apply(
                        lambda x: word in x.lower()
                    )
                )
                if pn_history_count != 0:
                    annotation_rate = annotation_count / pn_history_count
                    if annotation_rate > threshold:
                        annotation_rate_dic[feature_num].append(
                            {
                                "word": word,
                                "annotation_rate": annotation_rate,
                                "pn_history_count": pn_history_count,
                            }
                        )
#                 else:
#                     print(f"{feature_num}, {word}: pn_history_count=0")
    return annotation_rate_dic
```

```python
unigram_annotation_rate_word_list = match_annotation_rate(feature_num_list_unigram, threshold=0.99)
bigram_annotation_rate_word_list = match_annotation_rate(feature_num_list_bigram, threshold=0.99)
```
